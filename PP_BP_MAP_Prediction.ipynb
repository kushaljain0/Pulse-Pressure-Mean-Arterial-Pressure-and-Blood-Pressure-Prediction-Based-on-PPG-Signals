#### Training Model on Random Forest Regressor for PP

```python
from sklearn import ensemble
model = ensemble.RandomForestRegressor()
model = model.fit(X_train, y_train)
y_pred= model.predict(X_test)
```

#### R^2, MAE and MSE for PP

```python
from sklearn.metrics import r2_score
print(f"R^2 value : ", r2_score(y_test, y_pred))

from sklearn.metrics import mean_absolute_error
print(f"MAE value : ", mean_absolute_error(y_test, y_pred))

from sklearn.metrics import mean_squared_error
print(f"MSE value : ", mean_squared_error(y_test, y_pred))
```

```python
def best_fit(testY, testPredict):

    sty = 0
    for i in testY:
        sty += i[0]
    spre = 0
    for i in testPredict:
        spre += i
        
    
    xbar = sty/len(testY)
    ybar = spre/len(testPredict)
    n = len(testY) # or len(Y)

    numer = sum([xi[0]*yi for xi,yi in zip(testY, testPredict)]) - n * xbar * ybar
    denum = sum([xi[0]**2 for xi in testY]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    # print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

a, b = best_fit(y_test, y_pred)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred,color='Red')
yfit = [a + b * xi[0] for xi in y_test]
plt.plot(y_test, yfit,color ='black')
plt.xlabel("Target(mmHg)",fontsize=18)
plt.ylabel("PP Estimation(mmHg)",fontsize=18)
plt.legend(['R=0.989', f'Best fit Line y = {a} + x*{b}'],fontsize=18)
plt.savefig('PPscatter.jpg', format='jpg', dpi=1200)
```
